```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
plt.plot()
plt.title('My graph')
plt.xlim(0,10)
plt.show()
```

```python
plt.plot([1, 4, 3, 2])
```

```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
plt.plot(x, y)
```

```python
x = [1, 2, 3, 4]
y = [11, 22, 33, 44]
fig, ax = plt.subplots(figsize=(5,5))
ax.plot(x, y)
ax.set(title="Sample Simple Plot", xlabel="x-axis", ylabel="y-axis", xlim=(max(x),min(x)), ylim = (min(y),max(y)))
fig.savefig("my_plot.pdf")
```

```python
x = np.linspace(-4,4,100)
plt.plot(x,x**2)
plt.show()
```

```python
x = np.linspace(-4,4,10)
plt.scatter(x,x**2, color = 'blue', lw = 5)
plt.show()
```

```python
x = np.linspace(-4,4,101)
plt.hist(x)
plt.show()
```

```python
x = np.linspace(-4,4,101)
plt.bar(x,x**2)
plt.show()
```

```python
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2, 
                                             ncols=2, 
                                             figsize=(10, 5))

# Plot data to each axis
ax1.plot(x, x/2);
ax2.scatter(np.random.random(10), np.random.random(10));
ax3.bar(x, x);
ax4.hist(np.random.randn(1000));
```

```python
import scipy.stats as stats
x = stats.uniform.rvs(loc = 20, scale = 10, size = 10**4)
a = x.mean() - 3**0.5*x.std()
b = x.mean() + 3**0.5*x.std()
print(a,b)
print(x.min(),x.max())
```

```python
import seaborn as sns
x = stats.cauchy.rvs(loc = 0, scale = 1, size = 10**5)
x.mean()
sns.kdeplot(x)
```
