```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import warnings
warnings.filterwarnings('ignore')

# Load the "tips" dataset
tips = sns.load_dataset("tips")

tips
```

# Categorical Plots

- Categorical plots in data visualization are used to visualize categorical data, which consists of discrete and distinct categories or groups.

- The Count Plot displays the counts of observations in each category. Suitable for visualizing the frequency or distribution of categorical data.


```python
plt.style.use("fivethirtyeight")
plt.figure(figsize=(16,6))
chart=sns.countplot(x='day',data=tips,palette='Reds_r', order = tips.day.value_counts().index)
plt.title("Count of Tips by Day of the Week",
          fontsize='20',
          backgroundcolor='AliceBlue',
          color='red')
plt.show()
```

# Swarm Plots

- Swarm Plot typically has a categorical variable on the x-axis and a numerical variable on the y-axis, and it displays individual data points along each category. Now, As you can guess this gives us a visualization of the distribution and density of data points within categories.

- One of the main features of a Swarm Plot is that it minimizes overlap between data points. This means that each data point is plotted in such a way that it does not overlap with other points in the same category. This makes it easier to see the density of data. But it’s difficult to visualize with larger data, so it’s effective with relatively small datasets.

```python
plt.figure(figsize=(16,6))
sns.swarmplot(x="day", y="total_bill", data=tips, palette="Reds")
plt.title("Total Bill Distribution by Day of the Week",
          fontsize='20',
          backgroundcolor='AliceBlue',
          color='red')

plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.show()
```

```python
data = [5,8,12,18,19,19.9,20.1,21,24,28] 

fig, ax = plt.subplots()
sns.histplot(data, ax=ax)  # distplot is deprecate and replaced by histplot
ax.set_xlim(1,31)
ax.set_xticks(range(1,32,3))
plt.show()
```

# Point Plot

- The Point Plot typically has a categorical variable on the x-axis and a numerical variable on the y-axis and can provide insights into the distribution and central tendency of data within each category, such as a line or a point.


```python
plt.figure(figsize=(16,6))
sns.pointplot(x="day", y="total_bill", data=tips, ci=100, palette="Reds")
plt.title("Average Total Bill by Day of the Week",
          fontsize='20',
          backgroundcolor='AliceBlue',
          color='red')

plt.xlabel("Day of the Week")
plt.ylabel("Average Total Bill ($)")
plt.show()
```

# Categorical Box Plot

- A Categorical Box Plot, often referred to simply as a Box Plot or Box-and-Whisker Plot, is a type of categorical plot used to visualize the distribution of a numerical variable within different categories or groups, It displays quartiles (median, upper, and lower quartiles), potential outliers, and overall data spread.


```python
# Categorical Box Plot
plt.figure(figsize=(16, 6))
sns.boxplot(x="time", y="total_bill", data=tips, palette="Reds")
plt.title("Total Bill Distribution by Meal Time",
          fontsize='20',
          backgroundcolor='AliceBlue',
          color='red')

plt.xlabel("Meal Time")
plt.ylabel("Total Bill ($)")
plt.show()
```

# Categorical Violin Plot

- The categorical variable determines the groups or categories, and a numerical variable is used to create the Violin Plots within each category. The width of the plot at a specific point indicates the density of data points at that value, and the central part of a Violin Plot resembles a violin, which is why it’s called a Violin Plot.

- Inside the Violin, there is often a box plot representation that includes the median, quartiles, and potential outliers. Hence, This provides a summary of the central tendency and spread of the data to provide a comprehensive view of the distribution of a numerical variable within each category.

```python
# Categorical Box Plot
plt.figure(figsize=(16, 6))
sns.violinplot(x="time", y="total_bill", data=tips, palette="Reds")
plt.title("Total Bill Distribution by Meal Time",
          fontsize='20',
          backgroundcolor='AliceBlue',
          color='red')

plt.xlabel("Meal Time")
plt.ylabel("Total Bill ($)")
plt.show()
```

# CatPlot

- A Cat Plot, short for “Categorical Plot,” is a very powerful plotting function in Seaborn. It allows you to all the categorical plots we have seen above with just one parameter called **kind** in the command **sns.catplot(x,y)**.
- You can use kind= `bar`, `swarm` , `box`, `violin` , `count` , `point` , etc. Cat Plots work well for visualizing relationships between categorical variables and numerical variables. This is like a Swiss army knife for the categorical plots.

```python
# Categorical Box Plot
plt.figure(figsize=(16, 6))
sns.pointplot(x="time", y="total_bill", data=tips, palette="Reds")
plt.title("Average Bill Distribution by Meal Time",
          fontsize='20',
          backgroundcolor='AliceBlue',
          color='red')

plt.xlabel("Meal Time")
plt.ylabel("Total Bill ($)")
plt.show()
```

```python
#you can merage plots together
plt.figure(figsize=(16,6))
sns.swarmplot(x="day", y="total_bill", data=tips, color="#AF4343")
sns.boxplot(x="day", y="total_bill", data=tips, color="white")

plt.title("Total Bill Distribution by Day of the Week",
          fontsize='20',
          backgroundcolor='AliceBlue',
          color='red')

plt.xlabel("Day of the Week")
plt.ylabel("Total Bill ($)")
plt.show()
```

# KDE Plot

- A KDE (Kernel Density Estimation) Plot is a type of data visualization that is used to estimate the probability density function of a continuous variable. So, we will have our continuous variable on the x-axis and the y-axis represents the estimated probability density area under the curve sums to 1, indicating that the data distribution is normalized.

```python
plt.figure(figsize=(16, 6))
sns.kdeplot(data=tips['total_bill'], fill=True,color="#AF4343")
# or
#sns.kdeplot(x="total_bill",data=tips, fill=True,color="#AF4343")
plt.title("Kernel Density Estimation of Total Bill")
plt.xlabel("Total Bill ($)")
plt.ylabel("Density")
plt.show()
```

# Dist Plot

- A Dist Plot short for Distribution plot typically resembles a histogram, but additionally, it includes a smoothed curve, which is a KDE plot. It divides the range of the numerical variable into bins or intervals and displays the frequency or density of data points within each bin.



```python
plt.figure(figsize=(16,6))
sns.distplot(tips['total_bill'],color="#AF4343")
plt.title("Total Bill Distribution")
plt.xlabel("Total Bill ($)")
plt.show()
```

```python
plt.figure(figsize=(16,6))
sns.displot(tips['total_bill'],color="#AF4343")
plt.title("Total Bill Distribution")
plt.xlabel("Total Bill ($)")
plt.show()
```

# Strip Plot

- A Strip Plot is a bit similar to a Swarm Plot and displays individual data points along a single axis. However, the data points will overlap in the strip plot, so we can say that Swarm Plots are particularly useful with smaller datasets and when you want to prevent overlap.

```python
plt.figure(figsize=(16, 6))
sns.stripplot(x=tips['total_bill'], palette="Set1", jitter=True)
plt.title("Total Bill Distribution")
plt.xlabel("Total Bill ($)")
plt.show()
```

# Bivariate Plots<

- Bivariate plots involve the visualization and analysis of the relationship between two variables simultaneously. They are used to explore how two variables are related or correlated. 
- Common ones with Matplotlib are `sns.scatterplot(x,y,data)` , `sns.lineplot(x,y,data)` for scatter and line plots.

# Joint Plot

- A joint plot combines scatter plots, histograms, and density plots to visualize the relationship between two numerical variables. 
- The central element of a Joint Plot is a Scatter Plot that displays the data points of the two variables against each other, along the x-axis and y-axis of the Scatter Plot, there are histograms or Kernel Density Estimation (KDE) plots for each individual variable. These marginal plots show the distribution of each variable separately.
---
- Use `sns.jointplot(x,y,data=dataframe,kind)` , kind can be one of `[‘scatter’, ‘hist’, ‘hex’, ‘kde’, ‘reg’, ‘resid’]` these.


```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="scatter",color="#AF4343",palette="Reds",height=8)
plt.show();
```

# Hexbin Plot

- A Hexbin plot, short for Hexagonal Binning plot, groups data points into hexagonal bins, allowing you to visualize data density and patterns more effectively. These are especially valuable with large datasets when scatter plots with individual points become too crowded and hard to interpret!

- You can create a Hexbin plot, by using the kind parameter of joint plot. You can customize the color map, grid size, and other plot parameters to fine-tune the appearance of the Hexbin Plot.

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind="hex",color="#AF4343",palette="Reds",height=8)
plt.show();
```

# Multivariate Plots

- These are my favorite, these plots give us a lot of flexibility to explore the relationships and patterns among three or more variables simultaneously. 
- That is, Multivariate plots extend the analysis to more than two variables, which will be often needed in the Data Analysis.
- **Using Hue parameter:** Using the hue parameter will add color to the plot based on the provided categorical variable, specifying a unique color for each of the categories, This parameter can be used almost all of the plots like `.scatterplot()` , `.boxplot()` , `.violinplot()` , `.lineplot()` , etc.
- **Using size parameter:** size is another parameter that can be used to add another numeric variable dimension.


```python
fig,ax=plt.subplots(nrows=1,ncols=2,figsize=(16,8))
sns.violinplot(x="day",y="total_bill",data=tips, hue="sex",palette="Reds",
split=True ,      # Split violins by hue categories
ax=ax[0])

ax[0].set_title("Violin Plot with Hue for Total Bill by Day and Gender")
ax[0].set_xlabel("Day of the Week")
ax[0].set_ylabel("Total Bill ($)")
ax[0].legend(title="Gender")

# Scatter Plot with Hue and Size
plt.figure(figsize=(10, 8))
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="day",size="size", sizes=(20, 200), # Define the size range for markers
palette="Reds",ax=ax[1])

ax[1].set_title("Scatter Plot with Hue and Size for Tips Dataset")
ax[1].set_xlabel("Total Bill ($)")
ax[1].set_ylabel("Tip ($)")
ax[1].legend(title="Day")
plt.show();
```

# Relational Plot 

- A Relational Plot allows you to visualize the relationship between two numerical variables, along with additional categorical or numerical dimensions.

- Use `sns.relplot(x,y,data,hue,size,style)`. you can use the `hue` parameter to color data points by a categorical variable, the `size` parameter to vary marker size based on a numerical variable, and the `style` parameter to differentiate markers or lines by a categorical variable. The `kind` parameter allows you to specify the type of relational plot you want to create **scatter** plot or **line** plot.
- Assigning a `col`,`row` variables creates a faceted figure with multiple subplots arranged across the columns of the grid
- The size and shape of the plot is specified at the level of each subplot using the `height` and `aspect` parameters

```python
sns.relplot(data=tips, x="total_bill", y="tip", hue="day", col="sex",size="size",sizes=(20, 200),palette="Reds",height=6, aspect=1.2)
plt.show();
```

# Facet Grid 

- A Facet Grid is a feature in Seaborn that allows you to create a grid of subplots, each representing a different subset of your data. In this way, Facet Grids are used to compare patterns or relationships with multiple variables within different categories.
- To draw a plot on every facet, pass a function and the name of one or more columns in the dataframe to `FacetGrid.map()`
- The variable specification in `FacetGrid.map()` requires a positional argument mapping, but if the function has a **data** parameter and accepts named variable assignments, you can also use `FacetGrid.map_dataframe()`

```python
g=sns.FacetGrid(data=tips,  col="sex", height=6, aspect=1.2)
g=g.map_dataframe(sns.scatterplot,x="total_bill", y="tip", hue="day",size="size",sizes=(20, 200),palette="Reds")
plt.show();
```

```python
g = sns.FacetGrid(tips, col="time",  row="sex", height=4, aspect=1.7)
g.map_dataframe(sns.histplot, x="total_bill",kde=True,color="#AF4343")
plt.show()
```

- `sns.FacetGrid` and `sns.relplot` are both plotting functions in the Seaborn library, but they serve different purposes.

- `sns.FacetGrid` is used to create a grid of subplots based on a variable, allowing you to visualize relationships between multiple variables. It is a flexible function that can be used to create various types of plots such as scatter plots, line plots, bar plots, etc.

- On the other hand, `sns.relplot` is a higher-level interface that combines different types of relational plots, such as scatter plots and line plots, into a single plotting function. It is primarily used to visualize the relationship between two numeric variables.

- In summary, both `sns.FacetGrid` and `sns.relplot` have different functionalities. While FacetGrid is designed for creating subplots, relplot is designed for creating relational plots.

# Pair Plot 

- A Pair plot provides a grid of scatterplots, and histograms, where each plot shows the relationship between two variables, which is why it is also called a Pairwise Plot or Scatterplot Matrix.

- The diagonal cells typically display histograms or kernel density plots for individual variables, showing their distributions. The off-diagonal cells in the grid often display scatterplots, showing how two variables are related. Pair Plots are particularly useful for understanding patterns, correlations, and distributions across multiple dimensions in your data.

```python
"""
seaborn.pairplot(data, *, hue=None, hue_order=None, palette=None, vars=None, x_vars=None, y_vars=None, kind='scatter',
diag_kind='auto', markers=None, height=2.5, aspect=1, corner=False, dropna=False, plot_kws=None, diag_kws=None, grid_kws=None,
size=None)

kind:{‘scatter’, ‘kde’, ‘hist’, ‘reg’}
diag_kind:{‘auto’, ‘hist’, ‘kde’, None}
"""


penguins = sns.load_dataset("penguins")
sns.pairplot(penguins , kind='scatter',diag_kind='kde',hue="species"   ,palette="Reds",height=2.5, aspect=1.4)
plt.show()
```

```python
sns.pairplot(penguins , kind='hist',diag_kind='kde',hue="species"   ,palette="Reds",height=2.5, aspect=1.4,corner=True)
plt.show()
```

# Pair Grid

- By using a pair grid you can customize the lower, upper, and diagonal plots individually.

```python
g = sns.PairGrid(penguins ,hue="species" ,palette="Reds",height=2.5, aspect=1.4)
g.map_upper(sns.scatterplot)
g.map_diag(sns.kdeplot)
g.map_lower(sns.histplot)
g.add_legend()
plt.show()
```

# Heatmap

- Let’s start with the most commonly used plot, the Heatmap. Heatmap will give you an idea of how a variable is correlated with all other variables (numeric) in the data frame. We need to pass the correlation matrix to the heatmap and it will plot a graph with different colors, which will signify the correlation.



```python
import numpy as np
titanic = sns.load_dataset('titanic')
titanic=titanic.select_dtypes(include=np.number)
corr=titanic.corr()
mask=np.triu(np.ones_like(corr, dtype=bool))[1:,:-1]
corr=corr.iloc[1:,:-1].copy()
fig, ax = plt.subplots(figsize=(12,6))   
sns.heatmap(corr, mask=mask, vmin=-1, vmax=1, center=0, annot=True, fmt='.2f', 
            cmap='Reds', annot_kws={'fontsize':12,'fontweight':'bold'}, cbar=True)
ax.tick_params(left=False,bottom=False)
ax.set_xticklabels(ax.get_xticklabels(), rotation=45, horizontalalignment='right',fontsize=12)
ax.set_yticklabels(ax.get_yticklabels(), fontsize=12)
plt.title('Correlations between penguins Variables\n', fontsize=16)
fig.show()
```

# Cluster Map 

- A Cluster Map is a type of heatmap in Seaborn that not only displays data as a grid of colored cells but also arranges rows and columns in a way that groups similar data together, it uses hierarchical clustering algorithms to reorder rows and columns in the heatmap. To the side of the grouped heatmap, include dendrograms to make it a cluster map. Dendrograms are tree-like diagrams that show the hierarchical relationships between rows and columns.

```python
sns.clustermap(titanic.corr(), annot=True, cmap="Reds", figsize=(10, 6))
plt.title("Cluster Map of Iris Dataset Correlation")
plt.show()
```
